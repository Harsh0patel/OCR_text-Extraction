```python
from onnxruntime.transformers.models.gpt2.parity_check_helper import post_processing

from src import ocr_engine, preprocessing, utils
import time
import os
import re
```

```python
folder_path = ("C:/Users/hp333/Desktop/ocr/Data")
image_extension = (".jpg")

image_paths = []

for file in os.listdir(folder_path):
    if file.lower().endswith(image_extension):
        image_paths.append(os.path.join(folder_path, file))

print("Total images:", len(image_paths))
print(image_paths)
```

```python
engine = ocr_engine.PaddleOCREngine()
temp = 0
output_dir = "C:/Users/hp333/Desktop/ocr/results/output"
os.makedirs(output_dir, exist_ok=True)
valid_results = []

for image in image_paths:

    path = image
    processed_image = preprocessing.preprocess(path)
    print("preprocessing complete for: ", image)
    result = engine.predict(processed_image)
    print("prediction complete for: ", image)

    for res in result:
        temp += 1
        image_path = os.path.join(output_dir, f"result_img_{temp}.png")
        res.save_to_img(image_path)
        json_path = os.path.join(output_dir, f"result_json_{temp}.json")
        res.save_to_json(json_path)
        results = utils.text_from_json(json_path)
        # Expected pattern: [long alphanumeric]_[1-2 digits]_[2-5 letters]
        expected_pattern = r'^[0-9A-Z]{14,20}_\d{1,2}_[a-zA-Z]{2,5}$'
        for result in results:
            fixed_text = result['fixed']
            if fixed_text and re.match(expected_pattern, fixed_text, re.IGNORECASE):
                valid_results.append(result)
    time.sleep(2)
```
